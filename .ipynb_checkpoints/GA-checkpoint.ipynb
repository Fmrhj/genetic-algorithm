```python
def get_parents(pop, n_parents, fitness_function):
    pop.sort(key=lambda x: fitness_function(x[0], x[1]), reverse = True)
    return np.array(pop[:n_parents])
```

```python
size = 10
population = generate_initial_population(10, individual, ind_params)
```

```python
population
```

```python
# get 6 fittest parents
parents = get_parents(population, 6, fitness)
parents
```

```python
def crossover(parents, offspring_size):
    offspring = np.empty(offspring_size)
    crossover_point = np.uint8(offspring_size[0]/2)
    for k in range(offspring_size[1]):
        # Index of the first parent to mate.
         parent1_idx = k%parents.shape[0]
         # Index of the second parent to mate.
         parent2_idx = (k+1)%parents.shape[0]
         # The new offspring will have its first half of its genes taken from the first parent.
         offspring[k, 0:crossover_point] = parents[parent1_idx, 0:crossover_point]
         # The new offspring will have its second half of its genes taken from the second parent.
         offspring[k, crossover_point:] = parents[parent2_idx, crossover_point:]
    return offspring 
```

```python
offspring = crossover(parents, (2,2))
offspring
```

```python
def mutate(offspring_crossover, ind_params):
        for idx in range(offspring_crossover.shape[0]):
            # select randomly the gene where randomness is going to be added
            g = np.random.choice(range(offspring_crossover.shape[0]))

            # The random value to be added to the gene.
            offspring_crossover[idx][g] = offspring_crossover[idx][g] + \
                np.random.normal(0.2, 0.25, 1)

            if offspring_crossover[idx][g] > ind_params['upper_bound']:
                offspring_crossover[idx][g] = ind_params['upper_bound']

            if offspring_crossover[idx][g] < ind_params['lower_bound']:
                offspring_crossover[idx][g] = ind_params['lower_bound']

        return offspring_crossover
    
mutate_offspring = mutate(offspring, ind_params)
population = np.append(population, mutate_offspring, axis=0)
population.tolist().sort(key=lambda x: fitness(x[0], x[1]), reverse = True)
population = population[:size]
best_candidate = population[:1]
print(best_candidate)
fitness_values = [fitness(value_1, value_2) for value_1, value_2 in population.tolist()]
print(fitness_values[:1][0])
```

## Construct loop

```python
epochs = 1000
population = generate_initial_population(10, individual, ind_params)
for epoch in range(epochs):
    parents = get_parents(population, 6, fitness)
    offspring = crossover(parents, offspring_size=(2,2))
    offspring = mutate(offspring, ind_params)
    population.extend(offspring) 
    population.sort(key=lambda x: fitness(x[0], x[1]), reverse = True)
    population = population[:size]
    best_candidate = population[:1]
    score = fitness(best_candidate[0][0], best_candidate[0][1])
    
best_candidate, score
```

## Instatiate Evolution clas

```python
from ga import Individual, Population, Evolution, fitness

ind_params = {'lower_bound': -2,
              'upper_bound': 2,
              'number_of_genes': 2}

pop_params = {"n_parents": 6,
              "offspring_size":(2,2),
              "mutation_mean": 0.25,
              "mutation_sd": 0.5,
              'size': 10}

evo = Evolution(fitness, Population, Individual, pop_params, ind_params, fitness)
epochs = 200

history = []
for _ in range(epochs):
    evo.step()
    history.append(evo._best_score)
    
print(evo._best_individual, evo._best_score)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(history)
plt.title('Fitness of the best individual')
plt.xlabel('Epoch')
plt.ylabel('Fitness')
plt.savefig('public/fitness_history.png')
```
